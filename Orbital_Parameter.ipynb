```python
# =========================================
# WORKING orbital elements extraction
# =========================================

import numpy as np

if error is None:

    coords = orbits.coordinates

    # Extract values safely from PyArrow
    r = np.array([
        coords.x[0].as_py(),
        coords.y[0].as_py(),
        coords.z[0].as_py()
    ], dtype=float)

    v = np.array([
        coords.vx[0].as_py(),
        coords.vy[0].as_py(),
        coords.vz[0].as_py()
    ], dtype=float)

    # Solar gravitational parameter (AU^3/day^2)
    mu = 0.00029591220828559104

    r_norm = np.linalg.norm(r)
    v_norm = np.linalg.norm(v)

    # Angular momentum
    h = np.cross(r, v)
    h_norm = np.linalg.norm(h)

    # Inclination
    inc = np.degrees(np.arccos(h[2]/h_norm))

    # Node vector
    k = np.array([0.0,0.0,1.0])
    n = np.cross(k, h)
    n_norm = np.linalg.norm(n)

    # Eccentricity vector
    e_vec = (np.cross(v, h)/mu) - (r/r_norm)
    e = np.linalg.norm(e_vec)

    # Semi-major axis
    a = 1 / ((2/r_norm) - (v_norm**2/mu))

    # Longitude of ascending node
    Omega = np.degrees(np.arccos(n[0]/n_norm))
    if n[1] < 0:
        Omega = 360 - Omega

    # Argument of perihelion
    omega = np.degrees(np.arccos(np.dot(n, e_vec)/(n_norm*e)))
    if e_vec[2] < 0:
        omega = 360 - omega

    # True anomaly
    nu = np.degrees(np.arccos(np.dot(e_vec, r)/(e*r_norm)))
    if np.dot(r, v) < 0:
        nu = 360 - nu

    # Perihelion distance
    q = a * (1 - e)

    print("\n===== ORBITAL ELEMENTS =====\n")
    print("Semi-major axis a (AU):", a)
    print("Perihelion distance q (AU):", q)
    print("Eccentricity e:", e)
    print("Inclination i (deg):", inc)
    print("Longitude of Ascending Node Ω (deg):", Omega)
    print("Argument of Perihelion ω (deg):", omega)
    print("True anomaly ν (deg):", nu)

    print("\n===== OBSERVATION STATS =====")
    print("Rejected observations:", len(rejected_observations))
```

```python
# =========================================
# MPC STYLE OUTPUT FORMAT
# (Formatting only — no recalculation)
# =========================================

import numpy as np
from datetime import datetime

# ---- 1. Compute Epoch from observation data ----

dates = []

for line in obs_data_mpes.split("\n"):
    parts = line.split()
    if len(parts) > 3:
        try:
            year = int(parts[1][1:5])
            month = int(parts[2])
            day_frac = float(parts[3])

            day = int(day_frac)
            frac = day_frac - day

            dt = datetime(year, month, day)
            # This formula for JD is for a specific epoch, it might be more accurate to use astropy.time.Time
            # However, since the original code uses it, I'll keep it for consistency.
            jd = 2440587.5 + dt.timestamp()/86400.0 + frac

            dates.append(jd)
        except:
            # Handle cases where parsing fails for a line
            pass

epoch_jd = np.mean(dates) if dates else None # Handle case where dates might be empty

# ---- 2. z parameter ----

z_param = 1.0 / a

# ---- 3. Convert angles to radians ----

i_rad = np.radians(inc)
Omega_rad = np.radians(Omega)
omega_rad = np.radians(omega)
nu_rad = np.radians(nu)

# ---- 4. True anomaly to eccentric anomaly ----

# For hyperbolic orbits (e > 1), the formula for E changes.
# Given e is very close to 1 (0.999959784327443), it's likely nearly parabolic.
# The current formula assumes elliptic orbit. Need to verify if this is correct for e very close to 1.
# Assuming it's still elliptic as e < 1.
E = 2*np.arctan(np.tan(nu_rad/2) * np.sqrt((1-e)/(1+e)))

# ---- 5. Mean anomaly ----

M = E - e*np.sin(E)

# ---- 6. Mean motion ----

mu = 0.00029591220828559104 # Re-defining mu for self-containment, though it's in kernel state.
n_mean = np.sqrt(mu/(abs(a**3)))

# ---- 7. Perihelion time ----

T_peri = epoch_jd - (M/n_mean)

# ---- 8. P and Q vectors ----

P = np.array([
    np.cos(Omega_rad)*np.cos(omega_rad) - np.sin(Omega_rad)*np.sin(omega_rad)*np.cos(i_rad),
    np.sin(Omega_rad)*np.cos(omega_rad) + np.cos(Omega_rad)*np.sin(omega_rad)*np.cos(i_rad),
    np.sin(omega_rad)*np.sin(i_rad)
])

Q = np.array([
    -np.cos(Omega_rad)*np.sin(omega_rad) - np.sin(Omega_rad)*np.cos(omega_rad)*np.cos(i_rad),
    -np.sin(Omega_rad)*np.sin(omega_rad) + np.cos(Omega_rad)*np.cos(omega_rad)*np.cos(i_rad),
    np.cos(omega_rad)*np.sin(i_rad)
])


print("\n\nOrbital elements:")
print("    C/2026 A1 (MAPS)")

# Epoch
if isinstance(epoch_jd, (float,int)):
    print(f"Epoch JD TT = {epoch_jd:.1f}")
else:
    print("Epoch JD TT = UNKNOWN")

# Perihelion time
if isinstance(T_peri, (float,int)):
    print(f"T (Perihelion JD TT) {T_peri:.5f}")
else:
    print("T (Perihelion JD TT) UNKNOWN")

print()

print(f"q   {q:.7f}            (2000.0)            P               Q")
print(f"z  {z_param:+.7f}      Peri.  {omega:9.5f}     {P[0]:+0.8f}     {Q[0]:+0.8f}")
print(f"                 Node   {Omega:9.5f}     {P[1]:+0.8f}     {Q[1]:+0.8f}")
print(f"e   {e:.7f}      Incl.  {inc:9.5f}     {P[2]:+0.8f}     {Q[2]:+0.8f}")

print("\nFrom observations, rejected:", len(rejected_observations))
```
