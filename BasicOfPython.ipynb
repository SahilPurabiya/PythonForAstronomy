## Step 1: Preparing the Computational Tools


```python
import numpy as np
import matplotlib.pyplot as plt

u = 20                # here values are assigned to it.
theta = np.pi/4
g = 9.8
```

## Step 2 :From One Time Value to Many Time Values

```python
t = np.linspace(0, 5, 500)
```

## Step 3 : Evaluating the Equations at Every Time Step

```python
x = u * np.cos(theta) * t
y = u * np.sin(theta) * t - 0.5 * g * t**2
```

```python
# Keep only points where y >= 0
mask = y >= 0
x = x[mask]
y = y[mask]
```

## Step 4 : Visualizing the Motion

```python
plt.plot(x, y)
plt.xlabel("x position")
plt.ylabel("y position")
plt.title("Projectile Motion")
plt.grid()
plt.show()
```
