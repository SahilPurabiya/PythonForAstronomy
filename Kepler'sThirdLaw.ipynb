```python
!pip install astroquery pandas matplotlib numpy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astroquery.ipac.nexsci.nasa_exoplanet_archive import NasaExoplanetArchive
```

```python
table = NasaExoplanetArchive.query_criteria(
    table="pscomppars",
    select="pl_orbper,pl_orbsmax,st_mass",
    where="pl_orbper IS NOT NULL AND pl_orbsmax IS NOT NULL AND st_mass IS NOT NULL"
)

df = table.to_pandas()
df = df.dropna()

print("Number of systems:", len(df))
df.head()
```

```python
P_years = df["pl_orbper"] / 365.25
a_AU = df["pl_orbsmax"]
M_star = df["st_mass"]

lhs = P_years**2
rhs = a_AU**3 / M_star
```

```python
plt.figure(figsize=(7,6))

plt.scatter(np.log10(rhs), np.log10(lhs), s=5, alpha=0.5)

# 1:1 reference line
x = np.linspace(min(np.log10(rhs)), max(np.log10(rhs)), 100)
plt.plot(x, x, color='red')

plt.xlabel("log(a^3 / M*)")
plt.ylabel("log(P^2)")
plt.title("Testing Kepler's Third Law with Exoplanets")
plt.grid(True)
plt.show()
```

```python
coeff = np.polyfit(np.log10(rhs), np.log10(lhs), 1)
print("Slope:", coeff[0])
print("Intercept:", coeff[1])
```
