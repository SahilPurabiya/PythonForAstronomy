```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------
# INPUT DATA (ADD MULTIPLE GALAXIES)
# ---------------------------------

# Redshift values
z = np.array([
    0.023748,
    0.02836,
    0.006782,
    0.016266,
    0.03938,
    0.01818,
    0.016266,
    0.00298



])

# Distance values (Mpc)
distance = np.array([
    98.5,
    134.846,
    31.046,
    76.7,
    172,
    80.9,
    76.7,
    19.5


])

# ---------------------------------
# CONSTANT
# ---------------------------------

c = 299792  # km/s

# ---------------------------------
# CALCULATIONS
# ---------------------------------

# Calculate recessional velocities
velocity = z * c

# Individual Hubble constants
H0_individual = velocity / distance

# Average Hubble constant
H0_mean = np.mean(H0_individual)

print("------------ RESULTS ------------")
for i in range(len(z)):
    print(f"Galaxy {i+1}: H0 = {H0_individual[i]:.2f} km/s/Mpc")

print("----------------------------------")
print(f" Hubble Constant(slope) = {H0_mean:.2f} km/s/Mpc")

# ---------------------------------
# LINEAR FIT (BEST PRACTICE)
# ---------------------------------

# Fit line: velocity = H0_fit * distance
H0_fit, intercept = np.polyfit(distance, velocity, 1)

print(f"Hubble Constant from linear fit = {H0_fit:.2f} km/s/Mpc")

# ---------------------------------
# PLOT HUBBLE DIAGRAM
# ---------------------------------

distance_line = np.linspace(0, max(distance)*1.2, 200)
velocity_line = H0_fit * distance_line + intercept

plt.figure()

# Plot data points
plt.scatter(distance, velocity, label="Galaxies")

# Plot best-fit Hubble line
plt.plot(distance_line, velocity_line, label="Best Fit")

plt.xlabel("Distance (Mpc)")
plt.ylabel("Recessional Velocity (km/s)")
plt.title("Hubble Diagram (Multiple Galaxies)")

plt.legend()
plt.show()
```
